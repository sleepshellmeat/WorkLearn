```python
def load_data(path,index):
    #data_batch_1
    #batch:一批，10000个数据
    with open(path+str(index),'rb') as file:
        #将该二进制的file进行反序列化
        data = pickle.load(file,encoding='ISO-8859-1')
        x_data = data['data'].reshape((10000,3,32,32)).transpose((0,2,3,1))
        y_target = data['labels']
        return (x_data,y_target)
```

```python
def load_test_data(path):
    #data_batch_1
    #batch:一批，10000个数据
    with open(path,'rb') as file:
        #将该二进制的file进行反序列化
        data = pickle.load(file,encoding='ISO-8859-1')
        x_data = data['data'].reshape((10000,3,32,32)).transpose((0,2,3,1))
        y_target = data['labels']
        return (x_data,y_target)
```

```python
for i in range(1,6):
    if i == 1:
        x_data,y_target = load_data(path=path,index = i)
    else:
        data = load_data(path = path,index = i)
        x_data = np.concatenate([x_data,data[0]])
        y_target = np.concatenate([y_target,data[1]])
```

```python
x_test,y_test = load_test_data('./test_batch')
```

使用np.concatenate()方法级联所有训练数据

使用支持向量机作为机器学习模型

```python
from sklearn.svm import SVC
svc = SVC()
```

训练数据reshape

```python
X_train = x_data.reshape((50000,-1))
X_train.shape
```

使用PCA进行降维

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=80,whiten=True)
```

```python
pca.fit(X_train)

X_train_pca = pca.transform(X_train)
X_train_pca.shape
```

导入时间进行训练计时

```python
import time as time
```

进行训练

```python
print(time.time())
svc.fit(X_train_pca,y_target)
print(time.time())
```

获取测试数据

```python
#将预测的数据进行PCA降维
x_test2 = x_test.reshape((10000,-1))
x_test2_pca = pca.transform(x_test2)
```

```python
print(time.time())
y_ = svc.predict(x_test2_pca)
print(time.time())
```
