```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

### Series

#### 主要是为列表添加索引, <font color=red>Series传入的数据必须是1维的</font>

```python
nd = np.array([2,4,7,1])
```

```python
s = Series(nd)
s
```

```python
s.index = list('abcd')
s
```

```python
s1 = Series(nd,index=['张三', '李四', 'Otie', 'mark'])
s1
```

```python
s = Series(np.random.random(10), index = list('abcdefghig'))
s
```

```python
s['a']
```

#### Series 为1维的数组添加自定义的索引，自己封装了一个函数“.loc”但是原来的方法还是可以使用，这是显示索引

```python
s = Series([3, 2, 5, 9, 8], index = list('abcde'))
# index = list('')
```

```python
s.loc['c']
```

#### Series 还有其他添加索引的方法，并且“.iloc”为隐式索引

#### Series 在相加上面和NDarray有所不同

```python
s = Series([1, 2, None])
nd = np.array([1, 2, None])
```
