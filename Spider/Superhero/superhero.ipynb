```python
import pandas as pd
```

```python
human = pd.read_csv('marvel.csv')
```

### 将读取完的CSV数据生成DataFrame格式

```python
df = pd.DataFrame(human)
print(df)
```

### 一行一行的读取数据

```python
for i in range(len(df)):
    ducoment = df[i: i + 1]
    print(ducoment, '\n')
```

```python
for i in range(len(df)):
    combat = df['combat'][i]
    print(combat, '\n')
```

### read_CSV方法在读取文件中可以添加限制(<font color=red>这个只是在读取的时候添加的限制，用于开始之前的测试,</font>)

####  行限制nrows=100 只是取出100行

```python
person = pd.read_csv('marvel.csv', nrows=100)
df = pd.DataFrame(person)
print(df)
```

#### 列限制usecols=['name'，'combat', 'intelligence']，只是取出名字,耐力和智力

```python
person = pd.read_csv('marvel.csv', nrows=100, usecols=['name', 'combat', 'intelligence'])
df = pd.DataFrame(person)
print(df['name'])
```

### 失败的水平条形图

```python
import matplotlib
import matplotlib.pyplot as plt

ability = ['intelligence','combat','durability','power','speed','strength']
# 数组反转
ability.reverse()
# 取出英雄的6像数据
data = df[df['name']=='Iron Man'][['intelligence','combat','durability','power','speed','strength']]
# 绘图
fig, ax = plt.subplots()
b = ax.barh(range(6), data, color='#6699CC'))
# 为横向水平的柱图右侧添加数据标签。
for rect in b:
    w = rect.get_width()
    ax.text(w, rect.get_y()+rect.get_height()/2, '%d' %
            int(w), ha='left', va='center')
# 设置Y轴轴坐标上的刻度线标签
ax.set_yticks(range(6)))
ax.set_yticklabels(ability)
#不要X横坐标上的label标签。
plt.xticks(())

plt.title('水平横向的柱状图', loc='center', fontsize='25',
          fontweight='bold', color='red')
plt.show()
```
