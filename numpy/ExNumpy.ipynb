## ndarray的聚合操作

### max和min 能将数组里面的各种最大(小)值取出来，axis。

```python
cat = plt.imread('cat.jpg')
cat
```

```python
cat.max(axis=None) # 改变axis能取出各种最大值
```

```python
cat.min() # 最小值同理
```

```python
display(cat.max(), cat.min())
```

### mean 求平均值

```python
p2 = np.random.randint(0, 100, [3,4,4])
p2
```

```python
np.mean(p2, axis=None)
```

### sum求和

```python
np.sum(p2, axis=None)
```
